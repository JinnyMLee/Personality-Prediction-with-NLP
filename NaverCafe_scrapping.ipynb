```python
import requests
from bs4 import BeautifulSoup as bs
from selenium import webdriver
import time
import pandas as pd
import numpy as np
import csv
```

```python
#자신의 chromedriver가 있는 파일 위치를 path 변수에 할당
#window는 ~/chromedriver.exe를 불러오고 mac은 exe를 빼고 불러오기
driver = webdriver.Chrome(executable_path = "C:/Users/Jinny/Downloads/chromedriver.exe")
```

```python
# base_url = cafe main page url
base_url = 'https://cafe.naver.com/mbticafe/ArticleList.nhn?search.clubid=11856775'
driver.get(base_url)
driver.switch_to.frame('cafe_main')
```

```python
# 로그인하기
```

```python
# Create csv file
total_list = ['url', 'user', 'date', 'board', 'title', 'content', 'reply', 'view', 'like'] # column name in csv
f = open('mbti_cafe.csv', 'w', encoding = 'utf-8', newline='') # make new file if file doesn't exist
wr = csv.writer(f)
wr.writerow(total_list)
```

```python
cnt = 0  # number of collected data
page = 0 # position of current page
 
while page < 1000 : # Naver cafe's max page
    page = page + 1
    quest_urls = []
    # add personal conditions
    # &search.menuid = : 게시판 번호(카페마다 상이) -- &search.boardtype=L일 경우 전체 게시판
    # &search.page = : 데이터 수집 할 페이지 번호
    # &userDisplay = 50 : 한 페이지에 보여질 게시글 수
    driver.get(base_url + '&search.boardtype=L&search.page='+ str(page) +'&userDisplay=50')
    driver.switch_to.frame('cafe_main') #iframe으로 프레임 전환
    quest_list = driver.find_elements_by_css_selector('div.inner_list > a.article') #이 태그 부분 카페마다 다르다던데....
    quest_urls = [ i.get_attribute('href') for i in quest_list ]

    for quest in quest_urls :

        driver.get(quest)
        driver.switch_to.frame('cafe_main')
        
        if len(driver.find_elements_by_class_name("article-guide"))>0: # 등급이 낮아 못 읽는 게시판
            pass            
            continue
        
        else:
            # 현재 열려 있는 페이지를 parsing
            soup = bs(driver.page_source, 'lxml')  # 아니면 html.parser

            #제목
            try:
                title = soup.select('div.tit-box span.b')[0].get_text()
            except IndexError:
                title = ''

            #게시판
            try:
                board = soup.select('div.tit-box a')[0].get_text()
            except IndexError:
                board = ''

            #날짜
            try:
                date = soup.select('td.m-tcol-c.date')[0].get_text()
            except IndexError:
                date = ''

            #작성자
            if board == '':  # 간편게시판인 경우
                user = soup.select('div.tit strong')[0].get_text()
            else:
                user = soup.select('td.p-nick')[0].get_text()


            #내용
            content = soup.select('#tbody')[0].get_text().strip('\n').replace('\xa0','')

            # 댓글 수
            try:
                reply = soup.select('td.reply a')[0].get_text().strip('\n\t\t\t\t\t\t\t\n\t\t\t\t\t\t\t').lstrip('댓글 ')
            except IndexError:
                reply = ''

            # 조회 수
            try:
                view = soup.select('div.fl.reply_sort > table > tbody > tr span.b')[1].get_text()
            except IndexError:
                view = ''

            # 좋아요
            try:
                like = soup.select('div.fl.reply_sort em')[0].get_text()
            except IndexError:
                like =''

        # csv에 쓰기
        temp_list = [quest, user, date, board, title, content, reply, view, like]
        f = open('mbti_cafe.csv', 'a+', encoding = 'utf-8', newline='')
        wr = csv.writer(f)
        wr.writerow(temp_list)
        cnt = cnt + 1
        
    print([page, cnt]) #page로는 진행상황을 알 수 있고 cnt로는 누적으로 몇 개의 데이터를 모았는지 알 수 있음
    
f.close()
```
