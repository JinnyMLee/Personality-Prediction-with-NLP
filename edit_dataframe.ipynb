```python
import pandas as pd
import numpy as np
```

```python
first = pd.read_csv('mbti_cafe.csv')
second = pd.read_csv('mbti_cafe_sh.csv')
third = pd.read_csv('mbti_cafe(1-1000).csv')
```

```python
print(first.shape)
print(second.shape)
print(third.shape)
```

```python
del third['Unnamed: 0']
del third['type']
```

```python
# column 같으므로 concat 사용
concat = pd.concat([first, second, third])
concat.reset_index() # 인덱스 합친 다음 처음부터 다시 매겨주기
```

```python
28416+12332+48922
```

```python
concat.shape
```

```python
# 중복 제거
concat = concat.drop_duplicates(["title","content"])
concat = concat.drop_duplicates(["url"])
```

```python
concat.shape
```

```python
concat.head()
```

```python
# unique user count
concat['user'].unique().size
```

```python
# 게시판별 게시글 수 확인
concat.groupby('board').count()
```

```python
# 아이디 모두 소문자로 변환
concat['user'] = concat['user'].str.lower()
```

```python
# 새로운 column 추가
concat['type'] = np.nan
```

```python
# ID에 유형 포함된 경우 type에 기록해주기
# type: E/I x N/S x F/T x J/P
# 하드코딩.. 죄송합니다

concat.loc[concat['user'].str.contains('enfj'), 'type'] = 'enfj'
concat.loc[concat['user'].str.contains('enfp'), 'type'] = 'enfp'
concat.loc[concat['user'].str.contains('entj'), 'type'] = 'entj'
concat.loc[concat['user'].str.contains('entp'), 'type'] = 'entp'

concat.loc[concat['user'].str.contains('esfj'), 'type'] = 'esfj'
concat.loc[concat['user'].str.contains('esfp'), 'type'] = 'esfp'
concat.loc[concat['user'].str.contains('estj'), 'type'] = 'estj'
concat.loc[concat['user'].str.contains('estp'), 'type'] = 'estp'

concat.loc[concat['user'].str.contains('infj'), 'type'] = 'infj'
concat.loc[concat['user'].str.contains('infp'), 'type'] = 'infp'
concat.loc[concat['user'].str.contains('intj'), 'type'] = 'intj'
concat.loc[concat['user'].str.contains('intp'), 'type'] = 'intp'

concat.loc[concat['user'].str.contains('isfj'), 'type'] = 'isfj'
concat.loc[concat['user'].str.contains('isfp'), 'type'] = 'isfp'
concat.loc[concat['user'].str.contains('istj'), 'type'] = 'istj'
concat.loc[concat['user'].str.contains('istp'), 'type'] = 'istp'
```

```python
concat.head()
```

```python
# 유형별 게시글 수
concat.groupby('type')['user'].count().sort_values(ascending=False)
```

```python
# 유형별 unique user 수
concat.groupby('type')['user'].nunique().sort_values(ascending=False)
```

```python
concat.isnull().sum()
```

```python
concat.to_csv('mbti_cafe_total.csv')
```
